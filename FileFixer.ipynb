# Concatenate Files Together


```python
from PyPDF2 import PdfReader, PdfWriter
from google.colab import files

def combine_pdfs(paths, output_path):
    writer = PdfWriter()

    for path in sorted(paths):
        reader = PdfReader(path)
        for page in reader.pages:
            writer.add_page(page)

    with open(output_path, 'wb') as output_file:
        writer.write(output_file)

if __name__ == "__main__":
    # List of paths to PDF files to be combined
    pdfs = files.upload()
    # Output path for the combined PDF
    output_path = input('Enter PDF name') + ".pdf"


    combine_pdfs(pdfs, output_path)
    print("PDF files combined successfully!")
```

# Convert HEIC to png or jpg

```python
import os
import pyheif
from PIL import Image

def convert_heic_to_png(heic_file, output_dir):
    # Rename the HEIC file to HEIF since pyheif currently doesn't support HEIC directly
    renamed_file = os.path.splitext(heic_file)[0] + ".heic"
    os.rename(heic_file, renamed_file)

    # Open the HEIF file
    with open(renamed_file, "rb") as f:
        heif_image = pyheif.read(f)

        # Extract metadata
        metadata = heif_image.metadata

        # Extract the image data
        image = Image.frombytes(
            heif_image.mode,
            heif_image.size,
            heif_image.data,
            "raw",
            heif_image.mode,
            heif_image.stride,
        )

    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Save as PNG
    output_file = os.path.join(output_dir, os.path.splitext(os.path.basename(renamed_file))[0] + ".png")
    image.save(output_file, "PNG")

    print(f"Converted {renamed_file} to {output_file}")

    # Delete the renamed file
    os.remove(renamed_file)

# Example usage
if __name__ == "__main__":
     # Path to your HEIC file
    # print(str(heic_file))
    heic_file = files.upload()

    output_dir = "output"        # Output directory for PNGs
    convert_heic_to_png(heic_file, output_dir)
```

# Compile Clips Together

```python
from moviepy.editor import VideoFileClip, concatenate_videoclips
from google.colab import files
def combine_videos(video_paths, output_path):
    # Load all video clips
    video_clips = [VideoFileClip(path) for path in video_paths]

    # Concatenate the video clips
    final_clip = concatenate_videoclips(video_clips)

    # Write the final combined video to a file with MOV format
    final_clip.write_videofile(output_path, codec="libx264", fps=24)  # Adjust codec and fps as needed

    # Close all video clips
    for clip in video_clips:
        clip.close()

if __name__ == "__main__":
    # List of paths to input MOV files
    input_files = files.upload()

    # Path to the output combined MOV file
    output_file = "combined_output.mov"

    # Combine the videos
    combine_videos(input_files, output_file)
```

# Decrease video data size

```python
from moviepy.editor import VideoFileClip

def convert_mov_to_mp4(input_path, output_path):
    clip = VideoFileClip(input_path)
    clip.write_videofile(output_path, codec='libx264', audio_codec='aac')

# Example usage:
input_mov_path = files.upload()
input_mov_path = '/content/IMG_1307.MOV'
output_mp4_path = 'output_video.mp4'

convert_mov_to_mp4(input_mov_path, output_mp4_path)
print(f'{input_mov_path} has been converted to {output_mp4_path}')
```

```python
import pandas as pd

# Function to convert CSV to compressed pickle file
def csv_to_pkl(csv_file, pkl_file):
    # Read CSV file
    data = pd.read_csv(csv_file)
    # Save as compressed pickle file
    data.to_pickle(pkl_file, compression='gzip')
    print("CSV file converted to compressed pickle successfully!")

# Example usage
csv_file = 'data.csv'
pkl_file = 'data.pkl.gz'
csv_to_pkl(csv_file, pkl_file)
```
